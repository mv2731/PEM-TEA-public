```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ercot 2020 jop.csv', dtype= None)
df
```

```python
arr = np.array(df)
arr = arr.T
arr = arr[0:8]
arr
```

```python
def make_boxplot(title, dataDist, max):
    fig, ax = plt.subplots(figsize=(4, 8))
    plt.ylim(0,max)
    
    green_diamond = dict(markerfacecolor='g', marker='D')
    arr = dataDist
    
    my_dict = {'0':arr[0], '1':arr[1], '2':arr[2], 
              '3':arr[3], '4':arr[4], '5':arr[5], '6':arr[6], '7':arr[7]}
    
    ax.tick_params(axis='both', direction='in')
    ax.set_xticklabels(my_dict.keys(), color='k', fontname='Arial', fontsize=14)
    plt.xticks(color='k', fontname='Arial', fontsize=14)
    plt.yticks(color='k', fontname='Arial', fontsize=14)
    
    ax.set_ylabel('Optimal hourly current density (A cm$\mathregular{^{-2}}$)', fontname='Arial', fontsize=14)
    ax.set_xlabel('Average electricity price (\$ kWh$\mathregular{^{-1}}$)', fontname = 'Arial', fontsize=14)
    plt.title(title, fontname = 'Arial', fontsize=14)
        
    ax.boxplot(my_dict.values(), flierprops=green_diamond)
    print(my_dict.values())
    
    plt.savefig(title)
    
    plt.show()
```

```python
make_boxplot('Texas 2020 (Mode 3)', arr, 6.5)
```

```python
dfCali = pd.read_csv('JopOptimalAllYears_martinez2020.csv', dtype= None)
dfCali
```

```python
arrCali = np.array(dfCali)
arrCali = arrCali.T
arrCali = arrCali[1:9]
arrCali
```
